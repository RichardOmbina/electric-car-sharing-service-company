ANALYSIS

**Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.**

```
#loading our cleaned data set
df1=pd.read_csv('New.csv',index_col=0, encoding='utf-8')
df1
```

```
#
df2=df1[df1['CITY']=='Paris']
df2

```

```
df3=df2.groupby(['PUBLIC NAME', 'HOUR'])[['BLUECAR COUNTER']].sum()
df3
```

```
#calculating blue car usage by the hour
df3['BLUECAR_USAGE'] = df3['BLUECAR COUNTER'].diff()

df3.head()
```

```
#the negative values represent picked cars
df4 = df3[df3['BLUECAR_USAGE'] < 0]
df4.head()
```

```
#grouping our Bluecar dataset by hour
df5 = df4.groupby('HOUR')[['BLUECAR_USAGE']].sum()
df5
```

```
#sorting BLUECAR_USAGE by hour
df6=df5.sort_values('BLUECAR_USAGE', ascending=True)
df6
```

**1400hrs is the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.**
